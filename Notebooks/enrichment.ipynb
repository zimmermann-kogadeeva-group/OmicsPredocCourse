```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from enrichment import get_ora


sns.set_theme()
```

Load the DESeq2 results:

```python
df_lfc = pd.read_csv("../Output/DESeq2/dds_results.csv", dtype={"strain": str})
df_lfc
```

Get the GO terms for significantly up- or down-regulated genes:

```python
df_go_terms = (
    df_lfc
    .query("padj < 0.05 and abs(log2fc) > 1")
    .groupby(["strain", "comparison"])
    .gene_id
    # .apply(lambda x: get_ora(",".join(x)))
    .apply(get_ora)
    .reset_index()
    .drop(columns="level_2")
)
```

```python
df_go_terms.query("strain == '083.2' and comparison == 'CvsP' and fdr < 0.1")
```

```python
fig, ax = plt.subplots()
sns.barplot(
    df_go_terms
    .query("strain == '083.2' and comparison == 'CvsP' and fdr < 0.1")
    .sort_values("fdr"),
    y="term",
    x="number_in_list", 
    hue="fdr",
    palette="mako"
)
```
